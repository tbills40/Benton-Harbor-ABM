```python
#First upload edited files 
# load file to Read 
df = pd.read_csv('trippub.csv')
# create a trip ID for the dataframe
df['Person'] = df['HOUSEID'].map(str) + '-' + df['PERSONID'].map(str)


df1=df.loc[ :, ['HOUSEID','PERSONID','Person','HHFAMINC','HHSIZE','HHVEHCNT','LIF_CYC','HH_RACE','R_SEX','R_AGE','TRAVDAY','DWELTIME','TRPMILES','TRPTRANS','TRVLCMIN','URBRUR','WHYFROM','WHYTO','WTTRDFIN']]

df1
```

```python
df1["TRPTRANS"].replace({1:"Walk", 2: "Bicycle", 3:"Private car", 4:"Private car", 5:"Private car", 6:"Private car", 7:"Others", 8:"Others", 9:"Others", 10:"Others", 11:"Public bus", 12:"Others", 13:"Others", 14:"Others",15:"Others",16:"Others",17:"Others",18:"Others",19:"Others",20:"Others",97:"Others",-8:"Others",-7:"Others"},  inplace = True)
df1["LIF_CYC"].replace({1: "Single Individual", 2: "Couple", 3: "Single Parent", 4:"Couple with kids", 5:"Single Parent", 6:"Couple with kids", 7:"Single Parent", 8:"Couple with kids", 9:"Retired household", 10:"Retired household"}, inplace=True)
df1["WHYFROM"].replace({1: 'Home', 2: 'Home', 3:'Stop', 4:'Stop', 5: 'Stop', 6:'Stop', 7:'Stop', 8:'Stop', 9:'Stop', 10:'Stop', 11:'Stop', 12:'Stop',13:'Stop',14:'Stop',15:'Stop',16:'Stop',17:'Stop',18:'Stop',19:'Stop',97:'Stop',-7 :'Stop', -8 : 'Stop', 97:'Stop',-9:'Stop'}, inplace=True)
df1["WHYTO"].replace({1: 'Home', 2: 'Home', 3:'Stop', 4:'Stop', 5: 'Stop', 6:'Stop', 7:'Stop', 8:'Stop', 9:'Stop', 10:'Stop', 11:'Stop', 12:'Stop',13:'Stop',14:'Stop',15:'Stop',16:'Stop',17:'Stop',18:'Stop',19:'Stop',97:'Stop',-7 :'Stop', -8 : 'Stop', 97:'Stop',-9:'Stop'}, inplace=True)
df1["TRAVDAY"].replace({1: 'Weekend', 2: 'Weekday', 3:'Weekday', 4:'Weekday', 5: 'Weekday', 6:'Weekday', 7:'Weekend'}, inplace=True)
df1["R_SEX"].replace({1: 'Male', 2: 'Female', -8:"No", -7:"No"}, inplace=True)
df1["URBRUR"].replace({1: 'Urban', 2: 'Rural', -8:"No", -7:"No"}, inplace=True)
df1["HH_RACE"].replace({1:"White", 2: "Black", 3:'Asian', 4:'American Indian',5:'Native Hawaiian', 6:'Multiple responses selected', 97:'Some other race', -8:'Dont know' ,-7:'Refused'},  inplace = True)
df1["HHVEHCNT"].replace({0: "No",1:"Yes", 2: "Yes", 3:"Yes", 4:"Yes", 5:"Yes", 6:"Yes", 7:"Yes", 8:"Yes", 9:"Yes", 10:"Yes", 11:"Yes", 12:"Yes"},  inplace = True)

df1
```

```python
#categorization of variables
category = pd.cut(df1.HHFAMINC,bins=[0,3,7,90,10000],labels=['Poverty','Middleclass','Rich','Refused'])
df1.insert(5,'Incomeclass',category)
category = pd.cut(df1.TRPMILES,bins=[0,10,30,10000],labels=['Short Trip','Medium trip','Long TRip'])
df1.insert(5,'Triplenght',category)
category = pd.cut(df1.DWELTIME,bins=[0,240,10000],labels=['short Stop at ','Primary'])
df1.insert(5,'Primary Destination',category)
category = pd.cut(df1.R_AGE,bins=[0,5,17,65,99],labels=['Toddler/baby','Child','Adult','Elderly'])
df1.insert(5,'Ageclass',category)



df1
```
